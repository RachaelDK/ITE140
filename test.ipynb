*What is a Word Cloud?
It's a picture made of words where more frequent words apper bigger and bolder.

*Why Are They Useful?
They help people quickly see what's important in a large text( without having to read everything.)

*What is an N-Gram?

 * Unigram = Single word
 * Bigram = Two words in a row
 * Trigram = Three words in a row

Why Are N-Grams Useful?

They help us find common phrases in text.

Example: if lots of people say "Great product" in reviews, that tells us something positive about the product.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


text = "Christmas means presents and holiday movies. Families unwrap gifts and enjoy classics like Home Alone together."

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
nltk.download('punkt')
```

```python
from collections import Counter

text = "Artificial intelligence is transforming industries. It is changing how we work and live."

def tokenize(text):
    return text.lower().replace('.', '').replace(',', '').split()

tokens = tokenize(text)

bigram_list = [(tokens[i], tokens[i+1]) for i in range(len(tokens)-1)]

bigram_counts = Counter(bigram_list)


print(bigram_counts.most_common(3))
```
